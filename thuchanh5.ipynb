```python
#w5a10
def dang_cau(a, b):
    mp = {}
    mp2 = {}
    for i in range(len(a)):
        if a[i] in mp and mp[a[i]] != b[i]:
            return False
        if b[i] in mp2 and mp2[b[i]] != a[i]:
            return False
        mp[a[i]] = b[i]
        mp2[b[i]] = a[i]
    return True
a, b = input().split()
print(dang_cau(a,b))
```
